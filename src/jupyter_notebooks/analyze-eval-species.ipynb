```python
def plot_path_lengths(path_lengths_file, title="", ylabel="", xlabel="", out_file=None):
    df = pd.read_csv(path_lengths_file, sep='\t', header=None, names=['p1', 'p2', 'length'])
    # convert the inf to -1
    df.loc[df['length'] == float('inf'), 'length'] = -1
    # zoom in around < 4 or 5
#     df = df[(df['length'] <= 5) & (df['length'] > 0)]
    df = df[(df['length'] > 0)]
    df.plot.hist(by='length', bins=df['length'].nunique())
    
    plt.title(title)
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
#     plt.ylim(0,200)
    plt.xlim(0,5)
    plt.show()
    if out_file is not None:
        print("Writing figure to %s" % (out_file))
        plt.savefig(out_file)
    plt.close()
```

```python
# load the lengths file and plot a histogram
goid = "GO:0009432"
version = "2018_06-seq-sim-e0_1-string"
taxon="208964"
path_lengths_file = "outputs/viz/eval-species/distances/%s-%s-%s-%%s-nx.tsv" % (version, goid, taxon)
for comparison in ["pos-pos", "pos-neg", "neg-pos"]:
    plot_path_lengths(
        path_lengths_file % (comparison), 
        title="%s - %s - %s - %s" % (version, taxon, goid, comparison),
        ylabel="Frequency",
        xlabel="Shortest path length from left-out %s to %s of other species" % (comparison.split('-')[0], comparison.split('-')[1]),
    )
```
